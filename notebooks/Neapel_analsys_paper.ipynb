```python
import os 
import matplotlib.pyplot as plt 
import numpy as np
import pyvista
import cv2
import fnmatch
```

```python
dir = '../data/NEAPEL/Bildstapel_Thin_Lam_1_Seite_Kontrast/'

np_array = np.empty((2000,400,2116),dtype=np.uint8)

for n, file in enumerate(sorted(fnmatch.filter(os.listdir(dir),'*.jpg'))):
    image = np.array(cv2.imread(os.path.join(dir,file),cv2.IMREAD_GRAYSCALE))
    print(image.shape)
    np_array[:,:,n] = image
```

```python
np_array2 = np.transpose(np_array,axes=(1,0,2))
print(np_array2.shape,np_array2.dtype)
```

```python
plt.figure(figsize=(14,8))

nslice = 1800
image = np_array2[250,:,:]

rgb_image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

print(rgb_image.shape)

plt.imshow(rgb_image,cmap='gray')
```
